```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import LinearSVC

iris = datasets.load_iris()
X = iris["data"][:, (2,3)] # 花瓣長、寬
y = (iris["target"]==2).astype(np.float64)
```

```python
svm_clf = Pipeline((
    ("scaler", StandardScaler()),
    ("linear_svc", LinearSVC(C=1, loss="hinge")) # C 越小街道越寬
))

svm_clf.fit(X,y)
```

```python
svm_clf.predict([X[100]]) # SVM 分類器不會輸出每個類別的概率
```

### 添加特徵

衛星數據集

```python
X, y = datasets.make_moons(n_samples=100, noise=0.15, random_state=42)
```

```python
def plot_dataset(X,y):
    plt.plot(X[:,0][y==1],X[:,1][y==1],"bs")
    plt.plot(X[:,0][y==0],X[:,1][y==0],"g^")
    plt.xlabel(r"$x_1$", fontsize=20)
    plt.ylabel(r"$x_2$", fontsize=20, rotation=0)
```

```python
plot_dataset(X,y)
plt.show()
```

```python
def plot_predictions(clf, axes):
    x0s = np.linspace(axes[0], axes[1], 100)
    x1s = np.linspace(axes[2], axes[3], 100)
    x0, x1 = np.meshgrid(x0s, x1s)
    X = np.c_[x0.ravel(), x1.ravel()]
    y_pred = clf.predict(X).reshape(x0.shape)
    y_decision = clf.decision_function(X).reshape(x0.shape)
    plt.contourf(x0, x1, y_pred, cmap=plt.cm.brg, alpha=0.2)
    plt.contourf(x0, x1, y_decision, cmap=plt.cm.brg, alpha=0.1)
```

```python
from sklearn.preprocessing import PolynomialFeatures
polynomial_svm_clf = Pipeline([
    ("poly_features", PolynomialFeatures(degree=3)),
    ("scaler", StandardScaler()),
    ("svm_clf", LinearSVC(C=10, loss="hinge"))
])
polynomial_svm_clf.fit(X,y)
```

```python
plot_predictions(polynomial_svm_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y)
plt.show()
```

```python
from sklearn.svm import SVC
poly_kernel_svm_d3_r1_clf = Pipeline((
  ("Scaler", StandardScaler()),
  ("svm_clf", SVC(kernel="poly", degree=3, coef0=1, C=5))
))
poly_kernel_svm_d3_r1_clf.fit(X,y)

poly_kernel_svm_d10_r30_clf = Pipeline((
  ("Scaler", StandardScaler()),
  ("svm_clf", SVC(kernel="poly", degree=10, coef0=100, C=5))
))
poly_kernel_svm_d10_r30_clf.fit(X,y)
```

```python
plt.figure(figsize=(15,5))
plt.subplot(1,2,1)
plot_predictions(poly_kernel_svm_d3_r1_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y)
plt.subplot(1,2,2)
plot_predictions(poly_kernel_svm_d10_r30_clf, [-1.5, 2.5, -1, 1.5])
plot_dataset(X, y)
plt.show()
```

```python
# P.141, 142
rbf_kernel_svm_clf = Pipeline((
    ("scaler", StandardScaler()),
    ("svm_clf", SVC(kernel="rbf", gamma=5, C=0.001))
))

rbf_kernel_svm_clf.fit(X, y)
```

```python
hp = [
    [0.1,0.001],
    [0.1,1000],
    [5,0.001],
    [5,1000]
]

plt.figure(figsize=(15,10))
for i in range(4):
    rbf_kernel_svm_clf = Pipeline((
        ("scaler", StandardScaler()),
        ("svm_clf", SVC(kernel="rbf", gamma=hp[i][0], C=hp[i][1]))
    ))
    rbf_kernel_svm_clf.fit(X, y)
    plt.subplot(2,2,i+1)
    plt.title("gamma: {}, C:{}".format(hp[i][0],hp[i][1]))
    plot_dataset(X, y)
    plot_predictions(rbf_kernel_svm_clf, [-1.5, 2.5, -1, 1.5])

plt.show()
```
